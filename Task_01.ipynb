```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
population_df = pd.read_csv(r'C:\Users\mkk76\OneDrive\Documents\GitHub\SCT_DS_1\API_SP.POP.TOTL_DS2_en_csv_v2_87.csv', on_bad_lines='skip')
print(population_df.head())
```

```python
country_name = 'India'
country_data = population_df[population_df['Country Name'] == country_name]
```

```python
years = list(map(str, range(1960, 2024)))
population_data = country_data[years].values.flatten()
```

```python
plt.figure(figsize=(14, 8))
plt.plot(years, population_data, marker='o', color='b')
plt.title(f'Population Growth in {country_name} (1960-2023)')
plt.xlabel('Year')
plt.ylabel('Population')
plt.grid(True)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# histogram
plt.figure(figsize=(10, 5))
plt.hist(population_data, bins=10, color='blue', alpha=0.7)
plt.title(f'Population Distribution in {country_name} (1960-2023)')
plt.xlabel('Population')
plt.ylabel('Frequency')
plt.grid(True)
plt.tight_layout()
print('Histogram')
plt.show()
```

```python
# bar chart
plt.figure(figsize=(16, 8))
plt.bar(years, population_data, color='green', alpha=0.7)
plt.title(f'Population Growth in {country_name} (1960-2023)')
plt.xlabel('Year')
plt.ylabel('Population')
plt.xticks(rotation=45)
plt.grid(axis='y')
plt.tight_layout()
print('Bar Chart')
plt.show()
```
